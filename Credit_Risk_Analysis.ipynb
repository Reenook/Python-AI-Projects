Since the monthly income is null we can assume the individual applying for the loan is using something else as a collateral instead of showing an income to apply for a loan. based on this we can likely assume that most of the creditors are trying to hide their monlthy income or dependants status therefore we will fill it with 0 values because we do not have enough information.

```python
find_miss=train_redup[train_redup.NumberOfDependents.isnull()]
find_nmiss=train_redup[train_redup.NumberOfDependents.notnull()]
```

```python
find_miss.loc[:, 'NumberOfDependents'] = find_miss['NumberOfDependents'].fillna(0)
find_miss.loc[:, 'MonthlyIncome'] = find_miss['MonthlyIncome'].fillna(0)
```

```python
find_miss.isnull().sum()
```

```python
stats = find_nmiss['MonthlyIncome'].agg(['mean', 'median', 'std', 'min', 'max'])
stats_percent = stats / stats['max'] * 100
print(stats_percent)
```

```python
find_nmiss.loc[:,'MonthlyIncome']=find_nmiss['MonthlyIncome'].fillna(find_nmiss['MonthlyIncome'].median())
```

```python
find_nmiss.isnull().sum()
```

Now that we have gotten rid of the null values witin both of them we can combine them back together

```python
filled_train = pd.concat([find_miss, find_nmiss])
```
